### 1.2 Clean, wrangle, and handle missing data.

````python
# Identify columns with missing values
cols_with_missing = df.columns[df.isna().any()].tolist()
print("\nThe following columns contain missing values:\n",\
      cols_with_missing)
````

According to the data description in the 'README.txt', missing data in the ```Informational``` and ```PageValues``` columns might indicate users didn't interact with any informational pages make any purchase on the website, so replacing them with zero's seems like a reasonable decision. On the other hand, the ```OperatingSystems``` column appears to be a categorical column and we can assume that the missing data is random. In this case, a common method of dealing with missing data is **mode imputation**, which involves replacing the missing values with the most frequent category in the column. This method is less sensitive to outliers than using the mean:

````python
# Fill missing values with appropriate values
df['Informational'].fillna(0, inplace=True)
df['PageValues'].fillna(0, inplace=True)
df['OperatingSystems'].fillna(df['OperatingSystems']\
                              .mode()[0], inplace=True)

#https://vitalflux.com/pandas-impute-missing-values-mean-median-mode/
````

````python
# sum up all of the missing values in each row 
# to verify values were filled
print(df.isnull().sum())
````

**Normalization:**
Normalization is the process of scaling the features so that they have a similar range. There are different techniques for normalization, but we will use Min-Max scaling, which scales the values between 0 and 1 for the "duration" columns:

### 1.4 Reduce redundant data and perform need based discretization.
One potential area to look at is the ```BounceRates``` and ```ExitRates``` columns. These two columns are highly correlated, as both measure the rate at which users leave the website. Let's check the correlation coefficient between these two columns:

````python
df[['BounceRates', 'ExitRates']].corr()
````

As we can see, the correlation coefficient between ```BounceRates``` and ```ExitRates``` is quite high, at ```0.91```. This suggests that these two variables are measuring essentially the same thing, and we may be able to drop one of them without losing much information. For the purposes of this exercise, let's drop the ```BounceRates``` column:

````python
df = df.drop(columns=['BounceRates'])
df.head()
````

**Need-based discretization**

Now, let's perform need-based discretization on the ```ProductRelated_Duration``` column. We'll create three new columns: ```Low_Duration```, ```Medium_Duration```, and ```High_Duration```, which will indicate whether the duration is ```low``` (less than the 25th percentile), ```medium``` (between the 25th and 75th percentile), or ```high``` (greater than the 75th percentile):

````python
def duration_discretizer(x):
    if x < np.percentile(df['ProductRelated_Duration'], 25):
        return 'Low_Duration'
    elif x > np.percentile(df['ProductRelated_Duration'], 75):
        return 'High_Duration'
    else:
        return 'Medium_Duration'

df['ProductRelated_Duration_Disc'] = \
    df['ProductRelated_Duration'].\
    apply(duration_discretizer)
df = pd.get_dummies(df, \
                    columns=['ProductRelated_Duration_Disc'])
df.head()
````

#### 2.2 Provide measures of centrality and distribution with visualizations


````python
# Create a list of numerical variable columns only
numerical_vars = ['Administrative', 'Administrative_Duration', 'Informational', 
                  'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration',
                   'ExitRates', 'PageValues', 'Total_Duration']
                  
              
# Create subplot for each numerical column                   
fig, axs = plt.subplots(nrows=len(numerical_vars), figsize=(20, 30))


# Create the histograms for each column in the 'numerical_vars' list and set the title as the column name  

for i, col in enumerate(numerical_vars):
    axs[i].hist(df[col], bins=120, density=True)
    axs[i].set_title(col)

    # Add normal distribution curve to each histogram
    mu, std = norm.fit(df[col]) # Fit normal dist curve to column data
    xmin, xmax = axs[i].get_xlim() # Set the x-axis limits
    x = np.linspace(xmin, xmax, 100)
    p = norm.pdf(x, mu, std)
    axs[i].plot(x, p, 'c', linewidth=4)

    # Add red line at mean value
    axs[i].axvline(mu, color='r', linestyle='--', linewidth=4)
    
    
plt.tight_layout()
plt.show()

#https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html
#https://www.geeksforgeeks.org/matplotlib-axes-axes-hist-in-python/
#https://www.geeksforgeeks.org/how-to-plot-normal-distribution-over-histogram-in-python/
#https://www.geeksforgeeks.org/matplotlib-axes-axes-set_title-in-python/
#https://www.geeksforgeeks.org/matplotlib-axes-axes-get_xlim-in-python/
````

### 2.3 Diagnose for correlations between variables 


````python
# Calculate correlation matrix
corr_matrix = df[numerical_vars].corr()

# Visualize correlation matrix as heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix for Numerical Variables")
plt.show()
````

**Overall, we can see several strong and moderate correlations in the matrix, particularly between the ```duration``` variables (```Administrative_Duration```, ```Informational_Duration```, ```ProductRelated_Duration```, and ```Total_Duration```). These variables are likely highly correlated because ```Total_Duration``` is a product of our previous 'feature construction' and is comprised of the other three variables, so it is not surprising to see a strong correlation between them. Therefore, ```Total_Duration``` will need to be excluded from future regression analysis to avoid potential multicollinearity issues. Another solution is removing ```Administrative_Duration```, ```Informational_Duration```, and ```ProductRelated_Duration``` before conducting regression analysis to avoid multicollinearity and overfitting issues.**

**Continuing with the correlation analysis, ```Administrative_Duration``` and ```Administrative``` have a moderate positive correlation coefficient of 0.60. This correlation suggests that as the time spent on administrative pages increases, the number of administrative pages visited also tends to increase, and vice versa. Similarly, this correlation is not particularly surprising, as users who visit more administrative pages may need to spend more time on those pages to complete their tasks.
The correlation between ```Informational_Duration``` and ```Informational``` (0.62) is a moderate positive correlation, suggesting that as the time spent on informational pages increases, the number of informational pages visited tends to increase also. This correlation is also not particularly surprising, as users interested in a particular topic may spend more time exploring related informational pages.**
